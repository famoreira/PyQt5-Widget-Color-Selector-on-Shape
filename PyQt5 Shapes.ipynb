```python
import cv2
import numpy as np
import sys

from functools import partial
from PyQt5 import QtWidgets, uic
from PyQt5.QtWidgets import (QWidget, QSlider, QLineEdit, QLabel, QPushButton, QScrollArea,QApplication, QHBoxLayout, QVBoxLayout, QMainWindow)
from PyQt5.QtCore import Qt, QSize

#Color Codes
hex1 = '#7fc97f'
hex2 = '#beaed4'
hex3 = '#fdc086'
hex4 = '#ffff99'
hex5 = '#386cb0'
hex6 = '#f0027f'

#Color Names
name1 = 'Green'
name2 = 'Purple'
name3 = 'Orange'
name4 = 'Yellow'
name5 = 'Blue'
name6 = 'Fuchsia'

#Dimensions or "Geometry"
#bwidth and bheight must be greater then 20
bwidth = 30
bheight = 30
radius = bwidth/2
movex = bwidth - 20
movey = bheight - 20
ncol = 6

#Some UI options are availble here:
h_color = '#FFFFFF' #'#202020'
h_thick = 3

class Color_Picker(QWidget):
    def __init__(self):
        super().__init__()
        self.initUI()
    
    # Color buttons to select color for lips
    def initUI(self):
        #   .setGeometry( x,  y,  width,  height)
        self.setGeometry(30, 30, (ncol+1)*(bwidth + movex) - bwidth, 1*bheight + 2*movey)
        self.setWindowTitle("Color Selector")
        
        self.button1 = QPushButton(self)
        self.button1.setFixedWidth(bwidth)
        self.button1.setFixedHeight(bheight)
        self.button1.move(1*(bwidth + movex) - bwidth, movey)
        self.button1.setStyleSheet("*{background-color : " + str(hex1) + ";" + "color : " + str(hex1) + ";" + "border-radius: " + str(radius) + "px;}"
        + "*:hover{background: " + str(hex1) + ";" + "border: " + str(h_thick) + "px solid " + str(h_color) + ";" + "color: " + str(hex1) + ";}")
        self.button1.setText(str(name1))
        self.button1.clicked.connect(partial(self.clicked_btn, str(hex1)))

        self.button2 = QPushButton(self)
        self.button2.setFixedWidth(bwidth)
        self.button2.setFixedHeight(bheight)
        self.button2.move(2*(bwidth + movex) - bwidth, movey)
        self.button2.setStyleSheet("*{background-color : " + str(hex2) + ";" + "color : " + str(hex2) + ";" + "border-radius: " + str(radius) + "px;}"
        + "*:hover{background: " + str(hex2) + ";" + "border: " + str(h_thick) + "px solid " + str(h_color) + ";" + "color: " + str(hex2) + ";}")
        self.button2.setText(str(name2))
        self.button2.clicked.connect(partial(self.clicked_btn, str(hex2)))

        self.button3 = QPushButton(self)
        self.button3.setFixedWidth(bwidth)
        self.button3.setFixedHeight(bheight)
        self.button3.move(3*(bwidth + movex) - bwidth, movey)
        self.button3.setStyleSheet("*{background-color : " + str(hex3) + ";" + "color : " + str(hex3) + ";" + "border-radius: " + str(radius) + "px;}"
        + "*:hover{background: " + str(hex3) + ";" + "border: " + str(h_thick) + "px solid " + str(h_color) + ";" + "color: " + str(hex3) + ";}")
        self.button3.setText(str(name3))
        self.button3.clicked.connect(partial(self.clicked_btn, str(hex3)))

        self.button4 = QPushButton(self)
        self.button4.setFixedWidth(bwidth)
        self.button4.setFixedHeight(bheight)
        self.button4.move(4*(bwidth + movex) - bwidth, movey)
        self.button4.setStyleSheet("*{background-color : " + str(hex4) + ";" + "color : " + str(hex4) + ";" + "border-radius: " + str(radius) + "px;}"
        + "*:hover{background: " + str(hex4) + ";" + "border: " + str(h_thick) + "px solid " + str(h_color) + ";" + "color: " + str(hex4) + ";}")
        self.button4.setText(str(name4))
        self.button4.clicked.connect(partial(self.clicked_btn, str(hex4)))

        self.button5 = QPushButton(self)
        self.button5.setFixedWidth(bwidth)
        self.button5.setFixedHeight(bheight)
        self.button5.move(5*(bwidth + movex) - bwidth, movey)
        self.button5.setStyleSheet("*{background-color : " + str(hex5) + ";" + "color : " + str(hex5) + ";" + "border-radius: " + str(radius) + "px;}"
        + "*:hover{background: " + str(hex5) + ";" + "border: " + str(h_thick) + "px solid " + str(h_color) + ";" + "color: " + str(hex5) + ";}")
        self.button5.setText(str(name5))
        self.button5.clicked.connect(partial(self.clicked_btn, str(hex5)))

        self.button6 = QPushButton(self)
        self.button6.setFixedWidth(bwidth)
        self.button6.setFixedHeight(bheight)
        self.button6.move(6*(bwidth + movex) - bwidth, movey)
        self.button6.setStyleSheet("*{background-color : " + str(hex6) + ";" + "color : " + str(hex6) + ";" + "border-radius: " + str(radius) + "px;}"
        + "*:hover{background: " + str(hex6) + ";" + "border: " + str(h_thick) + "px solid " + str(h_color) + ";" + "color: " + str(hex6) + ";}")
        self.button6.setText(str(name6))
        self.button6.clicked.connect(partial(self.clicked_btn, str(hex6)))

        self.show()
        return

    # this causes the text for both color and color name to be accessible as a lipstick color
    def clicked_btn(self, value):
        sender = self.sender()
        # print(f'{sender.text()}')

        #the following removes the # from the hex so we can converst to BGR
        hex_code_data = f'{value}'
        hex_code_data = hex_code_data.replace("#", "")
        # print(hex_code_data)
        
        # this converts the hex w/o the # into BGR which is needed for the lip mask :)
        def hex_to_bgr(hex):
          rgb = []
          for i in (4, 2, 0):
            decimal = int(hex[i:i+2], 16)
            rgb.append(decimal)            
          return tuple(rgb)
        colors = hex_to_bgr(hex_code_data)
        # print(colors)
        
        ## ----- FOR SHAPE ----- ##
        # Create an empty image:
        m = 252
        r = 340
        src = np.zeros((r, m), dtype=np.uint8)

        # Read the empty image
        img = src
        img_rgb = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
        
        # x and y to move shape around
        xnum = 61
        ynum = 20
        # Define an array of endpoints of Hexagon
        points = np.array([[185 - xnum, 70 - ynum], [95 - xnum, 150 - ynum], [95 - xnum, 250 - ynum],
                           [185 - xnum, 330 - ynum], [275 - xnum, 250 - ynum], [275 - xnum, 150 - ynum]])

        # Use fillPoly() function and give input as image!
        cv2.fillPoly(img_rgb, pts=[points], color=colors)

        font = cv2.FONT_HERSHEY_SIMPLEX

        # Adds text to drawing based on f'{sender.text()}' 
        img_rgb_text = cv2.putText(img_rgb, f'{sender.text()}', (25, 35), font, 0.75, colors, 2, cv2.LINE_AA)

        # Displaying the image
        cv2.imshow("Shapes", img_rgb_text)

        # wait for the user to press any key to exit window
        cv2.waitKey(0)

        # Closing all open windows
        cv2.destroyAllWindows()

class SomeWidget(QWidget):
    def __init__(self, parent):
        super(QWidget, self).__init__(parent)
        
if __name__ == '__main__':
    app = QApplication(sys.argv)
    window = Color_Picker()
    sys.exit(app.exec_())
```
